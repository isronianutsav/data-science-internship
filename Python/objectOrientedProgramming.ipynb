```python
# heirarchical inheritence
class College:
    clg_name = "LNJPIT CHAPRA"

    def show_college(self):
        print("College:", self.clg_name)


class CSE(College):
    def show_details(self):
        print("Department: CSE")


class Civil(College):
    def show_details(self):
        print("Department: Civil")


class Mechanical(College):
    def show_details(self):
        print("Department: Mechanical")


# Object creation
cse = CSE()
civil = Civil()
mech = Mechanical()

# Method calls
cse.show_college()
cse.show_details()

civil.show_college()
civil.show_details()

mech.show_college()
mech.show_details()
```

POLYMORPHISM

```python
# ******************* Polymorphism (same function name)

def add(a, b):
    return a + b

def add(a, b, c=0):
    return a + b + c

print(add(2, 3))
print(add(2, 3, 4))
```

```python
# ******************* Polymorphism (Duck Typing)

class Cat:
    def sound(self):
        print("Cat says Meow")

class Dog:
    def sound(self):
        print("Dog says Bark")

def make_sound(animal):
    animal.sound()

cat = Cat()
dog = Dog()

make_sound(cat)
make_sound(dog)
```

```python
# ******************* Polymorphism (Method Overriding)

class Parent:
    def show(self):
        print("This is Parent class")

class Child(Parent):
    def show(self):
        print("This is Child class")

obj1 = Parent()
obj2 = Child()

obj1.show()
obj2.show() #Parent method is overridden


```

```python
print("super method application -- ")
#If you want to call the parent method from child, use super():

class Child2(Parent):
    def show(self):
        super().show()
        print("This is Child2 class")

obj3= Child2()
obj3.show() #Parent method is not overridden
```

ABSTRACTION

```python
#to hide  unnecessary things
# ******************* Abstraction using ABC

from abc import ABC, abstractmethod

class Vehicle(ABC): #You cannot create an object of Vehicle

    @abstractmethod
    def start(self):
        pass
    #pass means “no implementation here”

    #Forces child classes to define the method

```

```python
# ******************* Implementing abstract class
#Both classes override start()
class Bike(Vehicle):
    def start(self):
        print("Bike starts with kick")

class Car(Vehicle):
    def start(self):
        print("Car starts with key")

bike = Bike()
car = Car()

bike.start()
car.start()
```

```python

#abstraction + polymorphism
vehicles = [Bike(), Car()]

for v in vehicles:
    v.start()
```

Abstraction in Python is achieved using abstract base classes where abstract methods define a blueprint that child classes must implement.
