> ```with method```

````python
with open("demo.txt",'r') as file:
    print(file.read())#no need of file closing
````

> ERROR HANDLING WITH FILE HANDLING

````python
try:
    with open("demotext.txt",'r') as file:
        print(file.read())#no need of file closing
except FileNotFoundError as e:
    print(e)
````
