```python
try:
    pass
except:
    pass #works when try error
else:
    pass #works whenever try works
finally:
    pass #always work
```

```python
try:
    # Put the risky code here
    # This block is executed first
    # If an exception occurs here, Python immediately jumps to 'except'
    pass

except:
    # This block runs ONLY when an error occurs in the try block
    # If try runs successfully, this block is completely skipped
    pass

else:
    # This block runs ONLY when the try block executes without any error
    # It will NOT run if the except block has executed
    # Used for code that should run only in the success case
    pass

finally:
    # This block runs in EVERY situation
    # Runs whether an exception occurs or not
    # Commonly used for cleanup work like closing files or releasing resources
    pass
```
