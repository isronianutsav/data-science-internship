#### 2D ARRAY TO 1D ARRAY 
``` .ravel() , .flatten() ```

````python
arr = np.arange(60).reshape(12,5)
arr
````

````python
arr.ravel()
````

```ravel()```
- Returns a 1-D view of the original array (if possible)
- No new memory allocation (most of the time)
- Changes in result may affect original array

````python
arr.flatten()
````

```flatten()```
- Returns a 1-D copy
- Uses extra memory
- Changes do NOT affect original array

````python
#Bonus way
arr.reshape(-1)
````

##### `reshape(-1)` flattens an array into one dimension by automatically inferring the required size.
