````python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y, s = sizes, c = c , alpha = 0.5 )
plt.grid()       # adds grid lines to the plot
plt.colorbar()   # displays a color scale (legend) for color-mapped data
plt.show()
````

> HORIZONTAL BAR GRAPH

````python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.barh(x,y, color = c , height=0.5 ,alpha=0.6)  #horizontal graph with height,not width
plt.show()
````

> PIE CHART

````python

x = np.array([6, 4, 7, 8])
labels = np.array(['CO2', 'O2', 'N2', 'He'])
explode = [0, 0.2, 0, 0]   # use list/tuple, not set , explode extracts the slice
colors = ['green', 'pink', 'blue', 'red']

plt.pie(
    x,
    labels=labels,
    explode=explode,     # pulls out the selected slice
    shadow=True,         # adds shadow effect
    colors=colors,       # sets slice colors
    autopct="%1.1f%%",   # displays percentage with 1 decimal
    startangle=180       # rotates chart; first slice starts at 180
)


plt.show()
````

````python
x =np.array([1,2,3,4])
labels=np.array(['Co2','O2',"N2","He"])
explod = np.array([0,0.2,0,0])
color = np.array(['green', "pink","blue", "red"])

plt.pie(x, labels=labels, explode=explod,shadow=True, colors=color,autopct="%1.1f%%",startangle=90) #green wala 90degree
plt.legend(title="pie chart") #show colour indication 
plt.show()
````
