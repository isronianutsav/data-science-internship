> Now divide data into two parts (dependent and independent variables)

```python
x=df.drop('satisfaction',axis=1) #independent variables, after dropping eveything except price and storing in x variable
```

```python
y= df[['satisfaction']] #dependent variable (price left in df)
```

> ### DATA SPLITTING & SVC ALGORITHM

```python
from sklearn.svm import SVC 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#split data into further training and testing data
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape #for rows,columns
```

```python
svm=SVC() #creating object/model
svm.fit(x_train,y_train) #give training data to train the model
```

```python
import joblib #joblib used to save machine learning models
joblib.dump(svm,"model.lb") #svm is object and model.lb is model name
```

```python
model=joblib.load('model.lb') 
```

```python
model
```

```python
print(svm.score(x_train,y_train)) #here checking accuracy as it is a classification problem
print(svm.score(x_test,y_test))
```

```python
svm.predict(x_test) #returns the predicted class labels for the test data.
```

```python
model.predict(x_test)
```

```python
y_test #so exactly accuracy here happens
```
